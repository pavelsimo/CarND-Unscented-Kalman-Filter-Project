```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cmake-build-debug/ukf.csv')
nis_lidar = df[df['sensor_type'] == 'L']
x2_lidar = 5.991 # chi-squared 2 degrees of freedom, p-value 0.05 

nis_radar = df[df['sensor_type'] == 'R']
x2_radar = 7.815 # chi-squared 3 degrees of freedom, p-value 0.05 
```

```python
nis_lidar.head()
```

```python
nis_radar.head()
```

```python
print('Number of lidar measurements %d' % len(nis_lidar))
print('Number of radar measurements %d' % len(nis_radar))
```

```python
ax = nis_radar.plot(title='Radar NIS', y='nis', figsize=(10,4))
xmin, xmax = ax.get_xlim()
ax.hlines(y=x2_radar, xmin=xmin, xmax=xmax, color='r')
```

```python
ax = nis_lidar.plot(title='Lidar NIS', y='nis', figsize=(10,4))
xmin, xmax = ax.get_xlim()
ax.hlines(y=x2_lidar, xmin=xmin, xmax=xmax, color='r')
```

```python
r = (len(nis_lidar[nis_lidar['nis'] > x2_lidar]) / len(nis_lidar)) * 100
print('lidar: %.2f%%' % r)
```

```python
r = (len(nis_radar[nis_radar['nis'] > x2_radar]) / len(nis_radar)) * 100
print('radar: %.2f%%' % r)
```
